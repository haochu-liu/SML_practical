```python
# Import libraries
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
import numpy as np
import tensorflow as tf
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import Adam
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.utils import to_categorical
```

## Load dataset

```python
# Load the training data and the test inputs
X_train = pd.read_csv('X_train.csv', index_col = 0, header=[0, 1, 2]) # inputs of the training set
y_train = pd.read_csv('y_train.csv', index_col = 0, squeeze=True).to_numpy() # outputs of the training set
X_test =  pd.read_csv('X_test.csv', index_col = 0, header=[0, 1, 2]) # inputs of the test set
```

```python
# X_train is a 6,000 * 518 dataframe.
# Entries (i,j) correspond to the j'th dimension of the observation i
X_train
```

```python
number_classes=np.unique(y_train)
len(number_classes)
```

```python
#y_train contains the true class:  Electronic, Experimental, Folk, Hip-Hop, Instrumental, International, Pop or Rock
y_train
```

```python
# X_test is the array of test inputs, of the same format as X_train. The objective is to predict the class (Electronic, Experimental, Folk, Hip-Hop, Instrumental, International, Pop or Rock) of the output
X_test
```

## Neural Network

```python
from sklearn.preprocessing import LabelEncoder

# Initialize the LabelEncoder
label_encoder = LabelEncoder()

# Fit label encoder and return encoded labels
y_train_encoded = label_encoder.fit_transform(y_train)

# Now y_train_encoded contains integers representing the categories
```

```python
# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_train)

# Convert labels to one-hot encoding
y_onehot = to_categorical(y_train_encoded)

# Split the dataset into training and validation sets
X_training, X_val, y_training, y_val = train_test_split(X_scaled, y_onehot, test_size=0.2, random_state=42)
```

# Try a NN model first

```python
model = Sequential([
    Dense(512, activation='relu', input_shape=(518,)),  # Adjust input_shape based on your feature count
    Dropout(0.5),  # Helps prevent overfitting
    Dense(256, activation='relu'),
    Dropout(0.5),
    Dense(256, activation='relu'),
    Dropout(0.5),
    Dense(256, activation='relu'),
    Dropout(0.5),
    Dense(128, activation='relu'),
    Dense(8, activation='softmax')  # Output layer: one neuron per class, softmax for multi-class classification
])
```

```python
model.compile(optimizer=Adam(learning_rate=0.001),
              loss='categorical_crossentropy',  # Suitable for multi-class classification
              metrics=['accuracy'])
```

```python
history = model.fit(X_training, y_training,
                    validation_data=(X_val, y_val),
                    epochs=50,
                    batch_size=64)  # Tune based on your hardware capabilities
```

```python
# predict the y labels for X_test
y_pred=model.predict(X_test)
```

## NN tuning parameters

```python
from tensorflow.keras.layers import Input
from tensorflow.keras.models import Model

def create_model(n_layers=3, activation='relu'):
    inputs = Input(shape=(518,))
    x = inputs
    for _ in range(n_layers):
        x = Dense(256, activation=activation)(x)
        x = Dropout(0.5)(x)
    outputs = Dense(8, activation='softmax')(x)  # Assuming 8 classes
    model = Model(inputs=inputs, outputs=outputs)
    model.compile(optimizer=Adam(learning_rate=0.001),
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])
    return model
```

```python
from sklearn.model_selection import KFold

n_splits = 5  # For example, 5-fold cross-validation
kf = KFold(n_splits=n_splits, shuffle=True, random_state=42)
```

```python
results = []

for train_index, val_index in kf.split(X_scaled):
    X_training, X_val = X_scaled[train_index], X_scaled[val_index]
    y_training, y_val = y_onehot[train_index], y_onehot[val_index]

    model = create_model(n_layers=4, activation='relu')  # Adjust parameters as needed

    history = model.fit(X_training, y_training, validation_data=(X_val, y_val), epochs=50, batch_size=64, verbose=0)  # Set verbose to 0 to reduce log output

    results.append(model.evaluate(X_val, y_val)[1])  # Append accuracy

average_accuracy = np.mean(results)
print(f'Average Accuracy across folds: {average_accuracy}')
```

```python

def build_model(hp):
    model = Sequential()
    model.add(Dense(units=hp.Int('units', min_value=32, max_value=512, step=32),
                    activation=hp.Choice('activation', values=['relu', 'softplus', 'tanh', 'sigmoid']),
                    input_shape=(518,)))
    model.add(Dropout(0.5))
    for i in range(hp.Int('layers', 1, 5)):
        model.add(Dense(units=hp.Int(f'units_{i}', min_value=32, max_value=512, step=32),
                        activation=hp.Choice(f'activation_{i}', values=['relu', 'softplus','tanh', 'sigmoid'])))
        model.add(Dropout(0.5))
    model.add(Dense(8, activation='softmax'))
    model.compile(optimizer="adam", loss='categorical_crossentropy', metrics=['accuracy'])
    return model

tuner = RandomSearch(
    build_model,
    objective='val_accuracy',
    max_trials=5,  # Set smaller during experimentation
    executions_per_trial=1,  # Increase for more robust results
    directory='my_dir',
    project_name='hparam_tuning'
)

tuner.search(X_training, y_training, epochs=10, validation_data=(X_val, y_val))  # Use a lower number of epochs for the search phase
```

```python
# Get the best model
best_model = tuner.get_best_models(num_models=1)[0]
best_model
```

```python
best_hps = tuner.get_best_hyperparameters(num_trials=1)[0]
best_model = build_model(best_hps)
best_model
```

```python
history = best_model.fit(X_training, y_training,
                    validation_data=(X_val, y_val),
                    epochs=20,
                    batch_size=64)
```

```python
from sklearn.model_selection import KFold
import numpy as np

# Define the number of splits
n_splits = 5
kf = KFold(n_splits=n_splits, shuffle=True, random_state=42)

# Prepare an empty list to store results
validation_scores = []

# Loop over each split
for train_index, val_index in kf.split(X_scaled):
    # Generate the training and validation sets
    X_train_fold, X_val_fold = X_scaled[train_index], X_scaled[val_index]
    y_train_fold, y_val_fold = y_onehot[train_index], y_onehot[val_index]

    # Build the model (make sure to recompile it each time to reset weights)
    model = build_model(best_hps)  # Assuming build_model and best_hps are defined as before

    # Train the model
    model.fit(X_train_fold, y_train_fold, epochs=10, batch_size=64, verbose=0)  # Adjust epochs, batch_size as needed

    # Evaluate the model on the validation set
    val_score = model.evaluate(X_val_fold, y_val_fold, verbose=0)
    validation_scores.append(val_score[1])  # Assuming index 1 is accuracy

# Calculate the average validation score
average_val_score = np.mean(validation_scores)
print(f'Average Validation Accuracy: {average_val_score}')
```
